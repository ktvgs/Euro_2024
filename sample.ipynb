```python
country_string = df1.iloc[0, 16].strip("[]").replace("\n", " ")

# Split by single quotes followed by spaces, ensuring multi-word countries remain intact
country_list = [item.strip("'") for item in re.findall(r"'(.*?)'", country_string)]

di['away_team_name']=country_list
```
